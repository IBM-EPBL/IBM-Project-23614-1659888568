DATA VISUALIZATION

```python
data.hist(bins = 30, figsize = (10, 10), color = 'y')
```

```python
corr_matrix = data.corr()
corr_matrix
```

```python
#plotting the correlation matrix as a heatmap

fig = plt.figure(figsize=(12,8))
sb.heatmap(corr_matrix,annot=True)
plt.show()
```

```python
#plotting data which have high correlation

sb.relplot(data=data,x="GRE Score",y="Chance of Admit ",hue="Research",
            palette="dark",alpha=0.7)
plt.title("GRE Score vs Chance of Admit")
plt.show()
sb.relplot(data=data,x="CGPA",y="Chance of Admit ",hue="Research",
            palette="flare",alpha=0.7)
plt.title("CGPA vs Chance of Admit")
plt.show()
sb.relplot(data=data,x="TOEFL Score",y="Chance of Admit ",hue="Research",
            palette="Set2",alpha=0.7)
plt.title("TOEFL Score vs Chance of Admit")
plt.show()
```

```python
sb.barplot(data=data,x="SOP",y="Chance of Admit ",
            palette="Set1",hue="Research")
plt.title("SOP vs Chance of Admit")
plt.show()
sb.barplot(data=data,x="LOR ",y="Chance of Admit ",
            palette="flare",hue="Research")
plt.title("LOR vs Chance of Admit")
plt.show()
```

```python
sb.barplot(data=data,x="University Rating",y="Chance of Admit ",
            palette="Set1")
plt.title("University Rating vs Chance of Admit")
plt.show()
```
